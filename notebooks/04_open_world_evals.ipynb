# Open-World Evaluations

Load saved run artifacts for CommunityForensics, VCT2, RAID, and ARIA and summarize open-world behavior.

```python
from pathlib import Path
import json

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
RUNS = ROOT / "outputs" / "runs"
run_dirs = sorted(path for path in RUNS.iterdir() if path.is_dir()) if RUNS.exists() else []
print([path.name for path in run_dirs])
```

```python
if not run_dirs:
    print("No evaluation artifacts found. Run the eval scripts first.")
else:
    run_dir = run_dirs[-1]
    for dataset in ["commfor", "vct2", "raid", "aria"]:
        path = run_dir / "eval" / dataset / "metrics.json"
        if path.exists():
            payload = json.loads(path.read_text())
            print(dataset, list(payload.keys()))
        else:
            print(dataset, "missing")
```
