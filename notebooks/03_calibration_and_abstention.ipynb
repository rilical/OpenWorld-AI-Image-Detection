# Calibration and Abstention

Inspect temperature scaling, conformal artifacts, and evaluation-mode summaries from saved runs.

```python
from pathlib import Path
import json

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
RUNS = ROOT / "outputs" / "runs"
run_dirs = sorted(path for path in RUNS.iterdir() if path.is_dir()) if RUNS.exists() else []
print([path.name for path in run_dirs])
```

```python
if not run_dirs:
    print("No run artifacts found. Calibrate a run first with scripts/calibrate_temperature.py and scripts/build_conformal.py.")
else:
    run_dir = run_dirs[-1]
    for rel in ["calibration/temperature.json", "calibration/conformal.json", "eval/commfor/metrics.json"]:
        path = run_dir / rel
        print(f"\\n[{rel}]\\n")
        print(path.read_text()[:1200] if path.exists() else "Missing")
```
