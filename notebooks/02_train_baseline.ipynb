# Train Baseline

Notebook companion to `scripts/train_baseline.py`. Use it to inspect artifacts after a run finishes.

```python
from pathlib import Path
import json

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

RUNS = ROOT / "outputs" / "runs"
run_dirs = sorted(path for path in RUNS.iterdir() if path.is_dir()) if RUNS.exists() else []
print("Available runs:", [path.name for path in run_dirs])
```

```python
if not run_dirs:
    print("No training artifacts found. Run `python3 scripts/train_baseline.py --config configs/baseline_clip.yaml` first.")
else:
    run_dir = run_dirs[-1]
    print("Inspecting:", run_dir)
    print((run_dir / "train_summary.json").read_text() if (run_dir / "train_summary.json").exists() else "No train_summary.json yet")
```
